2) API de Rick e Morty

```python
import requests
import json
import pandas as pd

# Método que devolve os datos da resposta JSON
def descargar_pagina(url):
    response = requests.get(url)
    data=json.loads(json.dumps(response.json()))
    return data

data = descargar_pagina('https://rickandmortyapi.com/api/location')

# Creo un listado para guardar los datos
lista_datos = []
# Creo una lista para guardar los datos de cada elemento y después guardarlo en la lista general
lista_elemento = []
# Para guardar en cada caso que se requiera la url del residente
url_residente = ''

for i in data['results']:
    # Voy añadiendo todos los datos de cada elemento de la página
    lista_elemento.append(i['id'])
    lista_elemento.append(i['name'])
    lista_elemento.append(i['type'])
    lista_elemento.append(i['dimension'])
    # Recojo el primer enlace del primer residente de cada elemento (¡Si contiene algunos!)
    # Inicializo variables de las características de los residentes
    nombre_residente = ''
    especie_residente = ''
    if len(i['residents']) != 0:
        url_residente = i['residents'][0]
        v_residente = descargar_pagina(str(url_residente))
        nombre_residente = v_residente['name']
        especie_residente = v_residente['species']
        # Para poder añadir de la página de cada residente sus datos específicos
        lista_elemento.append(nombre_residente)
        lista_elemento.append(especie_residente)
        # Inicializo la url del residente para tenerla preparada para el siguiente
        v_residente = ''
    else:
        lista_elemento.append('')
        lista_elemento.append('')
    # En el caso de que no haya residentes, completa el espacio de la celta por nada
    # Agrego a la lista general los datos de cada elemento
    lista_datos.append(lista_elemento)
    # Inicializo la lista que tenemos de índice para cargar la lista general
    print(lista_elemento)
    # Vacío la lista de elementos para tenerla preparada para el siguiente.
    lista_elemento = []


len(lista_datos)
    
```
