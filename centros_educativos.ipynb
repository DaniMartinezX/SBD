```python
import os
import tempfile
```

> 2) Leer el archivo de los datos

```python
path_base = "C:/Users/daniel.martinezcarre/Downloads/"

dfCoruna = pd.read_csv(f"{path_base}provincia_corunha.csv")
dfLugo = pd.read_csv(path_base + "provincia_lugo.csv")
dfOurense = pd.read_csv(os.path.join(path_base, "provincia_ourense.csv"))
dfPontevedra = pd.read_csv(path_base+"provincia_pontevedra.csv")
```

Cómo combinar os DataFrames?

```python
dfTotal = pd.concat([dfCoruna,dfLugo,dfOurense,dfPontevedra])

dfTotal[["Concello"]]
```

Cómo gardar un arquivo temporal csv?

```python
arquivo_temporal = tempfile.mktemp()
dfTotal.to_csv(arquivo_temporal)
print(arquivo_temporal)
```

### JSON
